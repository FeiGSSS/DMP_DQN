```python
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
with open("score.npy", "rb") as f:
    score = pkl.load(f)
    
average = []
for i in range(len(score)):
    i0 = max(0, i-100)
    average.append(np.mean(score[i0:i]))
    
plt.figure(figsize=(20,10))
plt.scatter(list(range(len(score))), score)
plt.scatter(list(range(len(score))), average)
plt.grid(True)
plt.title("epoch score")
plt.xlabel("epoch")
plt.ylabel("Score")
plt.legend(["Epoch", "Average"])
plt.show()
```
